```python
from scipy import stats
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
np.random.seed(12345678)
```

```python
rvs1 = np.random.normal(0, 0.5, 1000)
rvs2 = np.random.normal(1, 0.9, 1000)
rvs3 = np.random.normal(5, 0.6, 1000)
rvs4 = np.random.normal(5, 0.7, 1000)
```

```python
sns.kdeplot(rvs1, shade = True, label ="rvs1")
sns.kdeplot(rvs2, shade = True, label ="rvs2")
sns.kdeplot(rvs3, shade = True, label = "rvs3")
sns.kdeplot(rvs4, shade = True, label = "rvs4")
```

```python
stats.ttest_ind(rvs1,rvs2)
```

```python
stats.ttest_ind(rvs2,rvs1)
```

```python
stats.ttest_ind(rvs2,rvs3)
```

```python
stats.ttest_ind(rvs3,rvs4)
```
